# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus

````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
# TOOL 3 : date and time
from datetime import datetime
def current_date():
    return datetime.now().strftime("%d-%m-%Y")
def current_time():
    return datetime.now().strftime("%H:%M:%S")
````

````python
# TOOL 4 : case  conventer
def change_case(text, mode):
    """
    Convert text to uppercase or lowercase.
    mode: 'upper' or 'lower'
    """
    try:
        if mode == "upper":
            return text.upper()
        elif mode == "lower":
            return text.lower()
        else:
            return "Invalid mode. Use 'upper' or 'lower'."
    except Exception:
        return "Error while changing case"
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

def agent(query: str):
    query_lower = query.lower()

    # TODO: Implement routing logic


    if "calculate" in query_lower:
      try:
        exp = query_lower.split("calculate")[1].strip()
        res = calculator(exp)
        return {
            "type" : "calculation",
            "result" : res
        }
      except Exception:
        return{
            "type" : "error",
            "result" : "Error in calculation"
        }

    elif "keywords" in query_lower:
      try:
        text = query_lower.replace("extract keywords", "")
        text = text.replace("keywords", "")
        text = text.replace("from", "").strip()
        res = extract_keywords(text)
        return {
            "type" : "keywords",
            "result" : res
        }
      except Exception:
        return{
            "type" : "error",
            "result" : "Error in extracting keywords"
        }

    elif "date" in query_lower:
      return {
        "type" : "general",
        "result" : current_date()
      }

    elif "time" in query_lower:
      return {
        "type" : "general",
        "result" : current_time()
      }
    elif "upper " in query_lower:
      try:
        text = query_lower.replace("upper", "")
        text = text.replace("case", "").strip()
        res = change_case(text, "upper")
        return {
            "type" : "general",
            "result" : res
        }
      except Exception:
        return{
            "type" : "error",
            "result" : "Error in changing case"
        }

    else:
      return {
        "type": "general",
        "result": "i am a simple agent assistant."
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?",
    "What is the date today?",
    "What is the time?",
    "upper case  what is lower letter?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
